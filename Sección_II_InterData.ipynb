##### Usarios Totales (2020-2022)

***Revista:*** Viendo los usuarios totales por sexo de estos últimos 3 años, podemos notar que el porcentaje de usuarios masculinos viene a la baja: pasando de ser un 47.5% del total de usuarios en 2020, al 47.1% en 2022

***Informe:*** Notemos que los datos están segmentados por sexo, pero queremos verlos totalizados por grupos de edad.
Si tenemos en mente que cada columna es un arreglo (unidimensional) de datos del mismo tipo, y en este caso su tipo de dato es numérica, podemos entenderlos fácilmente como vectores.

Para cada tabla por año, podemos ver al vector hombres (llamémoslo h) y al vector mujeres (llamémosle m).
Para poder ver un total por grupos de edad, simplemente podemos ejecutar una suma usual de vectores, donde cada uno de sus valores (o coordenadas, representadas por el grupo de edad correspondiente) se suman para darnos una resultante.

Esto mejor desarrollado de la siguiente manera:
 $\vec{total} = \vec{h_i} + \vec{m_i}$ donde i $\in$ {2020, 2021, 2022}

 De ahora en adelante cuando se haga una totalización de esta manera, recordaremos esta explicación

```python
# Nombramos a los vectores
h_2020 = uso_internet_2020[uso_internet_2020["Entidad Federativa"] == "CDMX"]["Hombres"]
h_2021 = uso_internet_2021[uso_internet_2021["Entidad Federativa"] == "CDMX"]["Hombres"]
h_2022 = uso_internet_2022[uso_internet_2022["Entidad Federativa"] == "CDMX"]["Hombres"]

m_2020 = uso_internet_2020[uso_internet_2020["Entidad Federativa"] == "CDMX"]["Mujeres"]
m_2021 = uso_internet_2021[uso_internet_2021["Entidad Federativa"] == "CDMX"]["Mujeres"]
m_2022 = uso_internet_2022[uso_internet_2022["Entidad Federativa"] == "CDMX"]["Mujeres"]

total_uso_internet_2020 = h_2020 + m_2020
total_uso_internet_2021 = h_2021 + m_2021
total_uso_internet_2022 = h_2022 + m_2022

# Indices
indice_edad = ["12-19 años", "20-29 años", "30-39 años", "40-49 años", "50-59 años", ">60 años"]

# Guardamos totales
total_uso_internet_2020 = pd.Series(list(total_uso_internet_2020), index = indice_edad)
total_uso_internet_2021 = pd.Series(list(total_uso_internet_2021), index = indice_edad)
total_uso_internet_2022 = pd.Series(list(total_uso_internet_2022), index = indice_edad)

totales_x_año_internet = pd.concat([total_uso_internet_2020, total_uso_internet_2021, total_uso_internet_2022], axis=1)
totales_x_año_internet.columns = ["2020", "2021", "2022"]
totales_x_año_internet
```

***Informe:*** Después de tener la tabla con los totales por grupo de edad, podemos totalizar ahora por año y tener una noción de cómo fue aumentando el uso de internet en cada año

```python
total_internet_2020 = totales_x_año_internet['2020'].sum(axis=0)
total_internet_2021 = totales_x_año_internet['2021'].sum(axis=0)
total_internet_2022 = totales_x_año_internet['2022'].sum(axis=0)

total_internet = pd.DataFrame([total_internet_2020, total_internet_2021, total_internet_2022], index=["2020", "2021", "2022"])
total_internet.columns = ["Total Usuarios"]
total_internet
```

```python
ax_1 = total_internet['Total Usuarios'].plot(kind='bar', figsize=(10, 6), title='Total Usuarios internet o smartphone por año')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.ylim(0, 7800000)
ax2 = plt.twinx()
ax2.plot(total_internet.values, linestyle='-', linewidth=2.0,color='red')


#Etiquetas para las barras
for bar in ax_1.patches:
        bar_value = bar.get_height()
        text = f'{bar_value:,.0f}'
        text_x = bar.get_x() + bar.get_width() / 2
        text_y = bar.get_y() + bar_value
        ax_1.text(text_x, text_y, text, ha='center',
          va='bottom', color="black",
          size=9, fontname = "serif")
plt.show()
```

***Informe:*** Tenemos dos segementos de recta que conforman una línea de tendencia. Ambos segementos corresponden al cambio que existió entre usuarios de internet totales en CDMX por año.

La pendiente de ambas rectas claramente es positiva, evidenciando un aumento de usuarios de internet entre 2020 y 2021, y 2021 y 2022. Sin embargo, también es claro que la pendiente del segmento correspondiente a los años 2021 y 2022 es menor, es decir, el segmento tiene una pendiente "menos pronunciada o más horizontal". Mostrando un aumento menor respecto al año anterior.



***Revista:*** La línea de tendencia en rojo, nos ayuda a visualizar de mejor manera los aumentos de usuarios de internet entre cada año.
```python
#Gráfica
ax4 = medidas_promedio['Promedio usuarios de cada medida (2020-2022)'].plot(kind='bar', title='Usuarios promedio por cada medida realizada (2020-2022)', figsize=(10,7), color=['blue','red','green','yellow','purple', 'black','brown'])
plt.gca().spines[['top', 'right',]].set_visible(False)

for bar in ax4.patches:
        bar_value = bar.get_height()
        text = f'{bar_value:,.0f}'
        text_x = bar.get_x() + bar.get_width() / 2
        text_y = bar.get_y() + bar_value
        ax4.text(text_x, text_y, text, ha='center',
          va='bottom', color="black",
          size=10)
plt.show()
```
